# PCA on Breast Cancer Data
In this notebook, we will apply Principal Component Analysis (PCA) on the breast cancer dataset to reduce its dimensionality. PCA helps in transforming high-dimensional data into a lower-dimensional space, while retaining as much variance as possible.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

### Load the Breast Cancer Dataset
We will load the breast cancer dataset, which contains 30 features for each sample. These features include measurements such as mean radius, mean texture, etc. The target variable represents whether a tumor is malignant or benign.

```python
# Load the breast cancer dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
X.head()
```

### Check the Shape of the Dataset
Let's examine the number of features and the number of samples in the dataset.

```python
# Check the shape of the dataset
X.shape, y.shape
```

### Standardize the Data
Since PCA is sensitive to the scale of the data, we need to standardize the features. This step ensures that each feature contributes equally to the analysis.

```python
# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

### Split the Data into Training and Test Sets
We will split the dataset into training and test sets (80% training and 20% testing) to evaluate the performance after applying PCA.

```python
# Split data into training and testing sets (80%/20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape
```

### Apply PCA with 2 Components
We will now reduce the dataset from 30 features to 2 principal components for visualization purposes.

```python
# Apply PCA with 2 components
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_pca[:5], X_test_pca[:5]
```

### Explained Variance Ratio
The explained variance ratio shows the proportion of the total variance in the data that is captured by each principal component. Let's check how much variance is explained by the first two components.

```python
# Explained variance ratio for the 2 principal components
pca.explained_variance_ratio_
```

### Visualizing the Explained Variance
We'll now visualize the explained variance ratio to understand how much of the dataset's variance is captured by the first two principal components.

```python
# Visualizing explained variance ratio for the 2 components
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.show()
```

### Visualize PCA Results (Training Set)
We will now visualize the PCA results for the training set in a 2D space.

```python
# Visualize PCA results for the training set
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='coolwarm', edgecolors='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Training Set (2D Projection)')
plt.colorbar()
plt.show()
```

### Visualize PCA Results (Test Set)
Next, we visualize the PCA results for the test set to see how well the principal components separate the data.

```python
# Visualize PCA results for the test set
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, cmap='coolwarm', edgecolors='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Test Set (2D Projection)')
plt.colorbar()
plt.show()
```

### Apply PCA with More Components
We will now apply PCA with 5 and 10 components to see how the explained variance and separation of data improve with more components.

```python
# Apply PCA with 5 components
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
pca.explained_variance_ratio_
```

### Visualize the Explained Variance for 5 Components
Letâ€™s visualize the explained variance ratio for the top 5 components.

```python
# Visualize the explained variance for 5 components
plt.figure(figsize=(8, 6))
plt.bar(range(1, 6), pca.explained_variance_ratio_)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance for Top 5 Components')
plt.show()
```

### PCA with 10 Components
We will now apply PCA with 10 components to further analyze the data and see how dimensionality reduction affects the variance and data distribution.

```python
# Apply PCA with 10 components
pca = PCA(n_components=10)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
pca.explained_variance_ratio_
```

### Visualize the Explained Variance for 10 Components
Letâ€™s visualize the explained variance ratio for the top 10 components.

```python
# Visualize the explained variance for 10 components
plt.figure(figsize=(8, 6))
plt.bar(range(1, 11), pca.explained_variance_ratio_)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance for Top 10 Components')
plt.show()
```

### Conclusion
Through this notebook, we demonstrated how PCA can reduce the dimensionality of the breast cancer dataset while retaining significant variance. We visualized the data in lower-dimensional spaces and examined how the number of principal components affects the explained variance and data separation.
