### Cell 1: Importing Libraries

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Cell 2: Loading the Dataset

```python
# Load the breast cancer dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

### Cell 3: Standardizing the Data

```python
# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Cell 4: Splitting the Data into Training and Test Sets

```python
# Split the data into training and test sets (80%/20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Cell 5: Applying PCA for Dimensionality Reduction

```python
# Apply PCA for dimensionality reduction - Automatically selecting the number of components to retain 95% of the variance
pca = PCA(n_components=0.95)  # Let PCA automatically choose the number of components;
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Check how many components are selected
print(f"Number of components selected by PCA: {X_train_pca.shape[1]}")
```

### Cell 6: Visualizing PCA Results (Training Set)

```python
# Visualize PCA results for the training set (first two components)
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='coolwarm', edgecolors='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Training Set (2D Projection)')
plt.colorbar()
plt.show()
```

### Cell 7: Training Random Forest Classifier

```python
# Train Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_pca, y_train)
```

### Cell 8: Predicting with Random Forest Classifier

```python
# Predict on the test set
y_pred_rf = rf_classifier.predict(X_test_pca)
```

### Cell 9: Evaluating Random Forest Classifier

```python
# Evaluate the performance of the Random Forest classifier
accuracy_rf = accuracy_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
report_rf = classification_report(y_test, y_pred_rf)

# Display Random Forest performance
print('Random Forest Classifier Performance:')
print(f'Accuracy: {accuracy_rf:.4f}')
print('Confusion Matrix:')
print(conf_matrix_rf)
print('Classification Report:')
print(report_rf)
```

### Cell 10: Training Decision Tree Classifier

```python
# Train Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train_pca, y_train)
```

### Cell 11: Predicting with Decision Tree Classifier

```python
# Predict on the test set
y_pred_dt = dt_classifier.predict(X_test_pca)
```

### Cell 12: Evaluating Decision Tree Classifier

```python
# Evaluate the performance of the Decision Tree classifier
accuracy_dt = accuracy_score(y_test, y_pred_dt)
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
report_dt = classification_report(y_test, y_pred_dt)

# Display Decision Tree performance
print('Decision Tree Classifier Performance:')
print(f'Accuracy: {accuracy_dt:.4f}')
print('Confusion Matrix:')
print(conf_matrix_dt)
print('Classification Report:')
print(report_dt)
```

### Cell 13: Comparing Classifier Performance

```python
# Compare the performances of both models
print('Performance Comparison:')
print(f'Random Forest Accuracy: {accuracy_rf:.4f}')
print(f'Decision Tree Accuracy: {accuracy_dt:.4f}')
```

### Cell 14: Conclusion of Classifier Comparison

### Conclusion
After reducing the dimensionality of the dataset using PCA, we trained two classifiers: Random Forest and Decision Tree. By comparing their performance in terms of accuracy and other evaluation metrics, we can conclude which model is better suited for breast cancer prediction.

### Cell 15: Exploring PCA Variance Explained

```python
# Display the variance explained by each principal component
print('Explained Variance Ratio by Each Component:')
print(pca.explained_variance_ratio_)
```

### Cell 16: Visualizing Explained Variance

```python
# Visualize the explained variance for each component
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.grid(True)
plt.show()
```

### Cell 17: Visualizing Confusion Matrix for Random Forest

```python
# Visualize the confusion matrix for Random Forest
import seaborn as sns

sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

### Cell 18: Visualizing Confusion Matrix for Decision Tree

```python
# Visualize the confusion matrix for Decision Tree
sns.heatmap(conf_matrix_dt, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.title('Confusion Matrix - Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

### Cell 19: Feature Importance from Random Forest

```python
# Visualize feature importance from Random Forest
importances = rf_classifier.feature_importances_
indices = np.argsort(importances)[::-1]

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.title('Feature Importance - Random Forest')
plt.bar(range(X_train_pca.shape[1]), importances[indices], align='center')
plt.xticks(range(X_train_pca.shape[1]), data.feature_names[indices], rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.show()
```
